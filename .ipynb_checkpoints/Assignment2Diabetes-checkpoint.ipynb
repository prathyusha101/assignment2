```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = '/Users/renukakandi/Desktop/Assignment2/diabetes.csv'
data = pd.read_csv(file_path)

# Set a seed for reproducibility
np.random.seed(42)

# Take a random sample of 25 observations
sample = data.sample(n=25)

# Calculate the mean and highest Glucose values for the sample
sample_mean_glucose = sample['Glucose'].mean()
sample_max_glucose = sample['Glucose'].max()

# Calculate the mean and highest Glucose values for the entire population
population_mean_glucose = data['Glucose'].mean()
population_max_glucose = data['Glucose'].max()
```

```python
# Print the results
print(f"Sample Mean Glucose: {sample_mean_glucose}")
print(f"Sample Max Glucose: {sample_max_glucose}")
print(f"Population Mean Glucose: {population_mean_glucose}")
print(f"Population Max Glucose: {population_max_glucose}")
```

```python
# Create charts to compare the statistics
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart for mean Glucose
ax[0].bar(['Sample', 'Population'], [sample_mean_glucose, population_mean_glucose], color=['blue', 'green'])
ax[0].set_title('Mean Glucose Comparison')
ax[0].set_ylabel('Mean Glucose')

# Bar chart for max Glucose
ax[1].bar(['Sample', 'Population'], [sample_max_glucose, population_max_glucose], color=['blue', 'green'])
ax[1].set_title('Max Glucose Comparison')
ax[1].set_ylabel('Max Glucose')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#b

# Calculate the 98th percentile of BMI for the sample
sample_98th_percentile_bmi = np.percentile(sample['BMI'], 98)

# Calculate the 98th percentile of BMI for the entire population
population_98th_percentile_bmi = np.percentile(data['BMI'], 98)

# Print the results
sample_98th_percentile_bmi, population_98th_percentile_bmi

# Create chart to compare the 98th percentile of BMI
fig, ax = plt.subplots(figsize=(6, 5))

# Bar chart for 98th percentile of BMI
ax.bar(['Sample', 'Population'], [sample_98th_percentile_bmi, population_98th_percentile_bmi], color=['red', 'green'])
ax.set_title('98th Percentile BMI Comparison')
ax.set_ylabel('98th Percentile BMI')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#c
# Bootstrap sampling to create 500 samples of 150 observations each
bootstrap_samples = [data.sample(n=150, replace=True) for _ in range(500)]

# Calculate statistics for BloodPressure in each bootstrap sample
bootstrap_means = [sample['BloodPressure'].mean() for sample in bootstrap_samples]
bootstrap_stds = [sample['BloodPressure'].std() for sample in bootstrap_samples]
bootstrap_percentiles = [np.percentile(sample['BloodPressure'], 98) for sample in bootstrap_samples]

# Calculate the average mean, standard deviation, and 98th percentile for the bootstrap samples
average_mean_bp = np.mean(bootstrap_means)
average_std_bp = np.mean(bootstrap_stds)
average_percentile_bp = np.mean(bootstrap_percentiles)

# Calculate the population statistics for BloodPressure
population_mean_bp = data['BloodPressure'].mean()
population_std_bp = data['BloodPressure'].std()
population_percentile_bp = np.percentile(data['BloodPressure'], 98)

# Print the results
average_mean_bp, average_std_bp, average_percentile_bp, population_mean_bp, population_std_bp, population_percentile_bp
```

```python
# Create charts to compare the statistics
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Bar chart for mean BloodPressure
ax[0].bar(['Bootstrap', 'Population'], [average_mean_bp, population_mean_bp], color=['blue', 'green'])
ax[0].set_title('Mean BloodPressure Comparison')
ax[0].set_ylabel('Mean BloodPressure')

# Bar chart for standard deviation of BloodPressure
ax[1].bar(['Bootstrap', 'Population'], [average_std_bp, population_std_bp], color=['blue', 'green'])
ax[1].set_title('Standard Deviation BloodPressure Comparison')
ax[1].set_ylabel('Standard Deviation BloodPressure')

# Bar chart for 98th percentile of BloodPressure
ax[2].bar(['Bootstrap', 'Population'], [average_percentile_bp, population_percentile_bp], color=['blue', 'green'])
ax[2].set_title('98th Percentile BloodPressure Comparison')
ax[2].set_ylabel('98th Percentile BloodPressure')

# Show the plots
plt.tight_layout()
plt.show()
```
